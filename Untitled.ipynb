```python
from test.factories import (
    CompanyFactory,
    CompanyNameFactory,
    LanguageFactory,
    TagFactory,
    UserFactory,
)

import pandas as pd
data = pd.read_csv('wanted_temp_data.csv')

for a in data.loc:
    index, values = a.index, a.values
    company_model = CompanyFactory()
    print(index, values)
    for i in range(len(index)):
        model, lang = index[i].split("_")
        print("11111",model, lang)
        language_model = LanguageFactory(name=lang)
        if values[i] != "nan" and model=="company":
            print("22222", values[i])
            company_name_model = CompanyNameFactory(name=values[i], c_id=company_model, language=language_model)
            print("33333", company_name_model)
        elif values[i] != "nan" and model=="tag":
            tags = values[i].split("|")
            for tag in tags:
                TagFactory(name=tag, language=language_model)
```
